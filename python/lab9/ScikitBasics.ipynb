**Zad. 7: Usu wiersze z brakujcymi wartociami i przypisz wynikowego data fame'a do `df_complete`. Wywietl pierwsze 10 wierszy i sprawd藕 czy znikn pity przykad.**

```python
df_complete = df_filtered.dropna()
df_complete.head(10)
```

**Zad. 8: Zamie Sex i Embarked na kolumny liczbowe.**

```python
# tu troch pomog...
# musimy zamieni (sowo klucz) wartoci tekstowe na liczby
# zr贸b co takiego dla kolumny Sex (zignoruj ostrze偶enie):
#
# df_complete.loc[:,"Sex"].replace({"female": 0, "male": 1}, inplace=True)
# df_complete.head(10)
#
# jeli to si uda - wymyl co podobnego dla kolumny Embarked

df_complete.loc[:,"Sex"].replace({"female": 0, "male": 1}, inplace=True)
df_complete.loc[:,"Embarked"].replace({"S": 0, "C": 1, "Q": 2}, inplace=True)
df_complete.head(10)

```

## Scikit-learn

Czas przej do uczenia maszynowego. Korzystajc z [dokumentacji scikit-learn](http://scikit-learn.org/stable/) oraz slajd贸w do zaj, spr贸buj wykona kolejne zadania. Polecam te偶 [tutoriale Kevina Markhama](http://blog.kaggle.com/author/kevin-markham/).

**Zad. 9: Korzystajc z scikit-learn podziel `df_complete` na zbi贸r uczcy i testowy, kolejno w proporcjach 60%-40%.**

```python
# to nie jest takie trudne, bo ju偶 zaimportowalimy odpowiedni funkcj (sp贸jrz na importy)
# doczytaj tylko co zwraca funkcja, kt贸rej u偶yjesz - scikit ma troch inne podejcie do zbior贸w danych ni偶 np. caret
# postaraj si, aby ten losowy podzia by powtarzalny...
# UWAGA: chcemy przewidywa warto zmiennej Survived
X = df_complete.drop(["Survived"], axis=1)
y = df_complete["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6, random_state=42)
```

**Zad. 10: Korzystajc ze zbioru treningowego zoptymalizuj parametry klasyfikatora RandomForestClassifier. U偶yj oceny krzy偶owej z miar AUC.**


```python
# sp贸jrz na slajdy do zaj, znajdujce si tam fragmenty kodu powinny Ci naprowadzi na rozwizanie
# uwa偶aj... bdziesz musia zaimportowa dodatkow bibliotek
# upewnij si te偶, 偶e Tw贸j eksperyment jest powtarzalny...

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=71)
rf.fit(X_train, y_train)
```

**Zad. 11: Oce uzyskany model na zbiorze testowym.**

```python
y_pred = rf.predict(X_test)

print(classification_report(y_test, y_pred))
print("Roc auc score:", roc_auc_score(y_test, y_pred))
```
